```python
class pessoa:
    def __init__(self,nome):
        self.nome = nome

class alunos(pessoa):
    def __init__(self,nome,matricula,id):
            super(). __init__(nome)
            self.matricula = matricula
            self.id = id

class professores(pessoa):
    def __init__(self,nome,id):
            super(). __init__(nome)
            self.id = id

class turma:
    def __init__(self, turma_alunos, ob_professor,nome_turma):
        self.turma_alunos = turma_alunos
        self.ob_professor = ob_professor
        self.nome_turma = nome_turma
        self.turma = []
        self.alunos = []
        self.professores = []

    def cadastro(self, aluno):
        self.alunos.append(aluno)

    def cadastro_prof(self,professor):
        self.professores.append(professor)

    def alunos_cadastrados(self):
        for aluno in self.alunos:
            print(f"Nome:{aluno.nome}, Matrícula:{aluno.matricula}, Turma: {aluno.id}")

    def professor_cadastrado(self):
        for professor in self.professores:
            print(f"Professor:{professor.nome}, Turma: {professor.id}")

turma_alunos = []
ob_professor = []
nome_turma = []
turmaA = turma(turma_alunos, ob_professor, nome_turma)
while True:
    print("=-"*17)
    print("|              MENU                |")
    print("| Opções:                          |")
    print("|","__"*16,"|")
    print("| 1- Cadastrar aluno               |")
    print("|","__"*16,"|")
    print("| 2- Cadastrar professor           |")
    print("|","__"*16,"|")
    print("| 3- Visualizar/sair               |")
    print("|","__"*16,"|")
    print("=-"*18)

    op = int(input("Escolha uma opção: "))
    if op == 1:
        nome = input("Nome do aluno: ")
        matricula = input("Número de matricula: ")
        id = input("Turma: ")

        aluno = alunos(nome,matricula,id)
        turmaA.cadastro(aluno)

    elif op == 2:
        print("=-"*17)
        print("|              MENU                |")
        print("| Opções:                          |")
        print("|","__"*16,"|")
        print("| 1- Cadastrar aluno               |")
        print("|","__"*16,"|")
        print("| 2- Cadastrar professor           |")
        print("|","__"*16,"|")
        print("| 3- Visualizar/sair               |")
        print("|","__"*16,"|")
        print("=-"*18)
        nome_professor = input("Nome professor: ")
        id = input("Turma: ")
        professor = professores(nome_professor,id)
        turmaA.cadastro_prof(professor)

    else:
        print("=-"*18,"\n")
        print("Alunos:")
        turmaA.alunos_cadastrados()
        print("__"*18,"\n")
        print("Professores:")
        turmaA.professor_cadastrado()
        print("__"*18,"\n")
        print("=-"*18)
        break

```
